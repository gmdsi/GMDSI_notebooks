# Downloading MODFLOW 6 and PEST++ Binaries

Before running any of the tutorials, you need the MODFLOW 6 and PEST++ executables. Both `flopy` and `pyemu` provide utility functions to download the latest releases directly from GitHub.

This notebook shows how to use these functions. You only need to run this once -- the downloaded binaries are cached in the repository's `bin/` folder and reused by all tutorials.

```python
import os
import flopy
import pyemu
```

## Set the target directory

We download all executables into a single `bin/` folder at the repository root. This is the same folder that the tutorial helper (`herebedragons.py`) uses when copying binaries into working directories.

```python
bin_dir = os.path.join("..", "..", "bin")
os.makedirs(bin_dir, exist_ok=True)
print(f"Target directory: {os.path.abspath(bin_dir)}")
```

## Download MODFLOW 6

`flopy.utils.get_modflow()` downloads the latest MODFLOW 6 and related USGS executables from the [MODFLOW-ORG/executables](https://github.com/MODFLOW-ORG/executables) GitHub repository. It automatically detects your operating system.

```python
flopy.utils.get_modflow(bin_dir)
```

## Download PEST++

`pyemu.utils.get_pestpp()` downloads the latest PEST++ suite from the [usgs/pestpp](https://github.com/usgs/pestpp) GitHub repository.

```python
pyemu.utils.get_pestpp(bin_dir, release_id="5.2.25")
```

## Verify

List the downloaded executables to confirm everything is in place:

```python
files = sorted(os.listdir(bin_dir))
print(f"{len(files)} executables downloaded:")
for f in files:
    print(f"  {f}")
```

## How the tutorials use these binaries

The tutorials call `hbd.prep_bins(working_dir)` to copy executables from `bin/` into each tutorial's working directory. If the `bin/` folder is empty, `prep_bins` will automatically trigger the downloads shown above -- so you don't strictly need to run this notebook first. But running it explicitly lets you verify that everything works before diving into the tutorials.
